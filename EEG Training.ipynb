```python
import os
import numpy as np
import mne
import matplotlib.pyplot as plt


sample_data_folder = mne.datasets.sample.data_path("/data/neuralabc/neuralabc_volunteers/Zaki/mne_practice/source")
sample_data_raw_file = os.path.join(sample_data_folder, 'MEG', 'sample','sample_audvis_raw.fif')
raw = mne.io.read_raw_fif(sample_data_raw_file) #To read a data file (change fif to the extension wanted)
raw.crop(0, 60).load_data()  # just use a fraction of data for speed here
```

# Artifacts.
   - Environmental artifacts: Brief signal jumps due to building vibration
   - Instrumentation artifacts: Continuous oscillations at specific frequencies used by head position indicator (HPI) coils.
   - Biological artifacts:
        - Short step-like deflections (especially in frontal EEG channels) due to eye movements
        - Large transient deflections (especially in frontal EEG channels) due to blinking
        - Brief bursts of high frequency fluctuations across several channels due to the muscular activity during swallowing

# Dealing with Artifacts.
1. Ignore the artifact and carry on with analysis.
2. Exclude the corrupted portion of the data and analyze the remaining data
3. Repair the artifact by suppressing artifactual part of the recording while (hopefully) leaving the signal of interest intact.

## Correction of distortions using a highpass filter
    -low-freqyency slow drifts could be seen at 20s and last around 20s.
    -to correct them, use a highpass filter higher than them in period (freq of 0.1Hz or 0.2Hz) 

```python
for cutoff in (0.1, 0.2):
    raw_highpass = raw.copy().filter(l_freq=cutoff, h_freq=None)
    fig = raw_highpass.plot(duration=60, order=mag_channels, proj=False,
                            n_channels=len(mag_channels), remove_dc=False)
    fig.subplots_adjust(top=0.9)
    fig.suptitle('High-pass filtered at {} Hz'.format(cutoff), size='xx-large',
                 weight='bold')
```

## Correction with ICA
 - when source signals are statistically independent and non-gaussian 
 - separate the sources using ICA, and then re-construct the sensor signals after excluding the sources that are unwanted.
 #### components could be selected manually/automatically (https://mne.tools/stable/auto_tutorials/preprocessing/plot_40_artifact_correction_ica.html#tut-artifact-ica)

```python
filt_raw = raw.copy()
filt_raw.load_data().filter(l_freq=1., h_freq=None)

ica = mne.preprocessing.ICA(n_components=25, random_state=97) #a very small number given that our data has over 300 channels, but with the advantage that it will run quickly and we will able to tell easily whether it worked or not. Could be run for all
ica.fit(filt_raw)
```

```python
raw.load_data()
ica.plot_sources(raw)
ica.plot_components()
ica.plot_properties(raw, picks=[0])
```

```python
# blinks before and after correction
ica.plot_overlay(raw, exclude=[0], picks='eeg')
```
